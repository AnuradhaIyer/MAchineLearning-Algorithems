```python
import numpy as np
```

```python
X=np.array([[1,5],
            [1,7],
            [2,6],
            [2,9],
            [3,3],
            [3,6],
            [3,8],
            [3,9],
            [4,5],
            [4,8],
            [5,4],
            [5,7],
            [6,9],
            [7,2],
            [7,3],
            [7,9],
            [8,1],
            [8,7]])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
from matplotlib import pyplot as plt

linked = linkage(X, 'single')

labelList = range(1, 19)

plt.figure(figsize=(2, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()  
```

```python
linked = linkage(X, 'average')

labelList = range(1, 19)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()  
```

```python
linked = linkage(X, 'complete')

labelList = range(1, 19)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()  
```
